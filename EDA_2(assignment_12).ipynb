handle missing values

```python
adult=adult.dropna()
adult
```

scaling techniques

```python
scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
```

```python
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
numerical_features
```

```python

adult_standard_scaled = adult.copy()
adult_minmax_scaled = adult.copy()
```

```python

adult_standard_scaled
adult_minmax_scaled
```

```python
adult_standard_scaled[numerical_features] = scaler.fit_transform(adult[numerical_features])
adult_minmax_scaled[numerical_features] = minmax_scaler.fit_transform(adult[numerical_features])
adult_standard_scaled[numerical_features]
adult_minmax_scaled[numerical_features]
```

encoding techniques

```python
encoder = OneHotEncoder(drop='first')
categorical_features = ['workclass', 'marital_status', 'relationship', 'race', 'sex']
encoder
categorical_features
```

```python
adult_encoded = pd.get_dummies(adult, columns=categorical_features)
adult_encoded
```

```python
label_encoder = LabelEncoder()
for col in ['occupation', 'native_country']:
    adult_encoded[col] = label_encoder.fit_transform(adult[col])
```

feature engineering

```python
adult['capital_net_gain'] = adult['capital_gain'] - adult['capital_loss']
adult['age_hours_per_week'] = adult['age'] * adult['hours_per_week']
adult['capital_net_gain']
adult['age_hours_per_week']
```

```python
adult['capital_gain_log'] = np.log1p(adult['capital_gain'])
adult['capital_gain_log']
```

feature selection

```python
iso_forest = IsolationForest(contamination=0.05)
outliers = iso_forest.fit_predict(adult[numerical_features])
adult= adult[outliers == 1]
iso_forest
outliers
adult
```
