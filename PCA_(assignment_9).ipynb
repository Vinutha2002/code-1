 3.investigate correlations between features

Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(wine)
```

Implement PCA and Determine the Optimal Number of Principal Components

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_scaled)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

Transform the Original Dataset into the Principal Components

```python

pca = PCA(n_components=0.9)
df_pca = pca.fit_transform(df_scaled)
pca
```

Clustering with Original Data

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans.fit_predict(wine)

silhouette_avg_original = silhouette_score(wine, clusters_original)
print(f'Silhouette Score (Original Data): {silhouette_avg_original}')
```

Visualize the Clustering Results

```python
plt.figure(figsize=(8, 6))
plt.scatter(wine.iloc[:, 0], wine.iloc[:, 1], c=clusters_original, cmap='viridis')
plt.title('Clustering Results (Original Data)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

 Clustering with PCA Data

```python

clusters_pca = kmeans.fit_predict(df_pca)

silhouette_avg_pca = silhouette_score(df_pca, clusters_pca)
print(f'Silhouette Score (PCA Data): {silhouette_avg_pca}')
```

Visualize the Clustering Results

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=clusters_pca, cmap='viridis')
plt.title('Clustering Results (PCA Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Comparison and Analysis

```python
print(f'Silhouette Score (Original Data): {silhouette_avg_original}')
print(f'Silhouette Score (PCA Data): {silhouette_avg_pca}')
```
